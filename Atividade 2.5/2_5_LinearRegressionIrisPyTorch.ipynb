# Regressão Linear com PyTorch

## Objetivo

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se
- o cálculo do gradiente automático do PyTorch, porém
- a função de perda e
- atualização dos parâmetros é feita de forma explícita.
- A rede é criada com uma camada nn.Linear()

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinear_Otimizacao.png)

## Importação dos pacotes

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn.datasets import load_iris

import torch
from torch import nn, optim
from torch.autograd import Variable

torch.manual_seed(1234)
```

## Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias
```

```python
print('x_train_bias.shape:',x_train_bias.shape, x_train_bias.dtype)
```

```python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
print(y_train)
```

## Model

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinearNeuronio.png)

```python
#Inicialização usando toch.nn.Linear

model = torch.nn.Linear(in_features=2, out_features=1, bias=False)
```

### Verificando a inicialização dos parâmetros

```python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight
```

### Testando o predict da rede

```python
xx = torch.ones((5,2))
xx
```

```python
model(Variable(xx))
```

## Treinamento

## Avaliação

```python
y_pred = model(Variable(x_train_bias))
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
```

### Acessando os parâmetros

```python
w_dic = model.state_dict()
w_dic
```

### Parâmetros ótimos, solução analítica

```python
x_bias = x_train_bias
y = y_train
x_bias_t = torch.t(x_bias)
w_opt = (torch.inverse(x_bias_t.mm(x_bias)).mm(x_bias_t)).mm(y)
print(torch.t(w_opt))
```

# Exercícios

- 1) Por que no final da varredura do laço de treinamento, a perda é impressa utilizando
   `loss.data`? Por que precisa do `data`?
   

> Resposta

* No PyTorch, loss.backward() calcula os gradientes para todos os tensores com requires_grad=True.
A variável loss é um tensor que guarda o histórico das operações para o cálculo do gradiente.
Usar loss.data pega apenas o valor numérico, sem esse histórico
útil para exibir a perda durante o treino sem afetar o gráfico de gradientes.
Se você usar loss direto, pode haver erros ou resultados inesperados, já que ele ainda contém as informações de gradiente.

- 2) Calcule e imprima o valor da perda (MSE) fazendo o predict da rede com os pesos já
   treinados.

```python
y_pred_trained = model(x_train_bias)
loss_trained = torch.mean((y_pred_trained - y_train)**2)

print('Perda de MSE com pesos treinados {:.6f}'.format(loss_trained.item()))
```

- 3) Crie duas listas `w0_list` e `w1_list` para guardar a evolução dos valores dos
   parâmetros durante o laço de treinamento.
   Modifique o exemplo do gráfico tipo **scatter plot** para visualizar a variação
   dos parâmetros do modelo duas dimensões.
   Marque neste gráfico o valor ótimo do parâmetro dado pela solução
   analítica.

```python
num_epochs = 100
learning_rate = 0.5

w0_list = []
w1_list = []

for epoch in range(num_epochs):
    inputs = x_train_bias
    target = y_train

    y_pred = inputs.mm(model.weight.t())

    loss = torch.mean((y_pred - target)**2)

    loss.backward()

    with torch.no_grad():
        model.weight.data -= learning_rate * model.weight.grad.data

    w0_list.append(model.weight.data[0][0].item())
    w1_list.append(model.weight.data[0][1].item())

    model.weight.grad.data.zero_()

    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.item()))
```

## Ploting vectors

```python
w_opt_0 = w_opt[0][0].item()
w_opt_1 = w_opt[1][0].item()

plt.figure(figsize=(8, 6))
plt.scatter(w0_list, w1_list, c=range(num_epochs), cmap='viridis', label='Evolução do parâmetro')
plt.scatter(w_opt_0, w_opt_1, color='red', marker='*', s=200, label='Melhores parâmetros')
plt.xlabel('w0 (Bias)')
plt.ylabel('w1 (Weight)')
plt.title('Evolução de parâmetros durante treinamento')
plt.legend()
plt.grid(True)
plt.show()
```

# Aprendizados

> Principais aprendizados do código e dos gráficos:

* Regressão Linear no PyTorch: Mostra como criar um modelo simples de regressão linear e atualizar os pesos manualmente, sem usar otimizador pronto.

* Autodiferenciação: Mesmo com atualizações manuais, o loss.backward() calcula automaticamente os gradientes — base do treino de redes neurais.

* Função de perda (MSE): Usa o erro quadrático médio, que mede a diferença entre valores previstos e reais.

* Evolução dos parâmetros: O gráfico mostra como w0 e w1 mudam ao longo das épocas para reduzir a perda.

* Solução analítica vs. treinada: Comparar os pesos treinados com os ideais mostra se o modelo convergiu bem.

* Taxa de aprendizado: Controla a velocidade e a estabilidade do treino — taxas altas podem acelerar ou desestabilizar a convergência.

* Normalização dos dados: Essencial para garantir que todas as variáveis influenciem igualmente o treino e evitem problemas de convergência.

Esses pontos resumem como treinar um modelo linear simples e entender conceitos básicos do PyTorch, como tensores, gradientes automáticos e atualização de parâmetros.
