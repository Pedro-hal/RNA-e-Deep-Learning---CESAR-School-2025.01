## Importação dos pacotes

```python
%matplotlib inline
import torch

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

torch.manual_seed(1234)

def print_assert(r):
    if r:
        return 'OK!'
    else:
        return 'Incorreto!'
```

## Dataset

### Leitura dos dados

```python
import pandas as pd
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_in = data[:,0:1]
y_in = data[:,1:2]

x_train = torch.FloatTensor(x_in)
y_train = torch.FloatTensor(y_in)

n_samples = x_train.size(0)
print('x_train.shape:',x_train.shape, type(x_train))
print('y_train.shape:',y_train.shape, type(y_train))

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])

iris_pd = pd.DataFrame(x_in, columns=['x_in'])
iris_pd['y_in'] = y_in
iris_pd.head()
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()

plt.scatter(x_train,y_train)
plt.xlabel('Comprimento sepalas')
plt.ylabel('Comprimento petalas')
```

### Exercício 1 - Criação do `x_train_bias`

faça o `x_train_bias` ser uma matriz com 1s na primeira coluna e x_train na segunda coluna.




Dica: você pode precisar da função [torch.cat](http://pytorch.org/docs/master/torch.html#torch.cat) e [torch.ones](http://pytorch.org/docs/master/torch.html#torch.ones).

```python

x_train_bias = torch.cat((torch.ones(n_samples, 1), x_train), dim=1)
print(x_train_bias.size())
```

```python

print(x_train_bias[:, 1].shape)
print(x_train.shape)
```

## Modelo da rede

### Definição da classe

```python
class Net():
    def __init__(self, n_in, n_out):
        self.w = torch.Tensor(n_out, n_in)
        self.w.uniform_(-0.1,0.1) #inicialização dos parâmetros

    def forward(self, x_bias):
        return x_bias.matmul(torch.t(self.w))
```

### Instanciando a rede

```python
model = Net(2,1)
```

### Testando o predict da rede

```python
model.forward(torch.ones((4,2)))
#model.forward(torch.randn(4,2))
```

## Treinamento

### Exercício 2 - Treinamento

Refaça o código de treinamento da rede, mas agora utilizando somente Tensores do PyTorch. Faça o mesmo procedimento que o notebook [LinearRegressionIrisNumPy](LinearRegressionIrisNumPy.ipynb). Para consultar as operações disponíveis no PyTorch use a [documentação do PyTorch](http://pytorch.org/docs/master/torch.html).

Funções que podem ser úteis:
- torch.t
- torch.mean
- torch.matmul

```python
torch.manual_seed(42)

n_features = x_train_bias.shape[1] # Use x_train_bias to get the correct number of features
w = torch.randn(n_features, 1, dtype=torch.float32, requires_grad=True)
b = torch.randn(1, dtype=torch.float32, requires_grad=True)

def model(X):
    return torch.matmul(X, w) + b

def mse_loss(y_pred, y_true):
    return torch.mean((y_pred - y_true) ** 2)

learning_rate = 0.01
epochs = 1000

for epoch in range(epochs):
    # Forward
    y_pred = model(x_train_bias) # Use x_train_bias in the model
    loss = mse_loss(y_pred, y_train)

    # Backward
    loss.backward()

    # Atualização manual dos parâmetros
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

    # Zerando os gradientes acumulados
    w.grad.zero_()
    b.grad.zero_()

    # Exibição periódica do erro
    if (epoch + 1) % 100 == 0:
        print(f'Época [{epoch+1}/{epochs}], Loss: {loss.item():.6f}')

print("Pesos finais:\n", w.detach().numpy())
print("Bias final:", b.item())
```

## Avaliação
